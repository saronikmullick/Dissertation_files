ULEZ statistics

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Date": ["2023-06", "2023-09", "2024-02", "2024-09"],
    "Unique_diesel_cars": [378100, 326300, 330900, 326400],
    "Noncompliant": [111500, 55300, 48000, 43200],
    "Compliant": [266600, 270900, 282900, 283200],
    "Pct_noncompliant": [29.5, 17.0, 14.5, 13.2],
    "Pct_compliant": [70.5, 83.0, 85.5, 86.8],
})

plt.figure(figsize=(9,5))
sns.lineplot(data=df, x="Date", y="Pct_compliant", marker="o", label="Compliant")
sns.lineplot(data=df, x="Date", y="Pct_noncompliant", marker="o", label="Non-compliant")
plt.title("London ULEZ: Diesel-car compliance over time")
plt.ylabel("Share of diesel cars (%)")
plt.xlabel("")
plt.ylim(0, 100)
for i, r in df.iterrows():
    plt.text(i, r["Pct_compliant"]+1, f"{r['Pct_compliant']:.1f}%", ha="center")
    plt.text(i, r["Pct_noncompliant"]-4, f"{r['Pct_noncompliant']:.1f}%", ha="center")
plt.legend(frameon=False)
plt.tight_layout()
plt.show()
```
